```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img=[]
for i in range(255):
    t=[]
    for j in range(1920):
        t.append([i,0,0])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([i,0,0])
    img.append(t)
for i in range(255):
    t=[]
    for j in range(1920):
        t.append([0,i,0])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([0,i,0])
    img.append(t)
for i in range(255):
    t=[]
    for j in range(1920):
        t.append([0,0,i])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([0,0,i])
    img.append(t)
    
img=np.array(img) 
plt.imshow(img)

cv.imwrite('qwe.png',img)
```

```python
img = cv.imread('rgb.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow("Image",img)
cv.waitKey(0)

img = cv.imread('Transition.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow("Image",img)
cv.waitKey(0)
```

```python
img = cv.imread('rgb.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

r = img[ :  , : , 0 ]  # - Red
g = img[ :  , : , 1 ]  # - Green
b = img[ :  , : , 2 ]  # - Blue


cv.imshow("Frame",r)
cv.waitKey(0)

cv.imshow("Frame",g)
cv.waitKey(0)

cv.imshow("Frame",b)
cv.waitKey(0)
```
