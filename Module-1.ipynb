## Lets Create a Chess Board

```python
zeros=np.zeros((8,8),dtype=int)
for i in range(len(zeros)):
    for j in range(len(zeros[i])):
        if((i+j)%2)==0:
            zeros[i][j]=1
plt.imshow(zeros)
```

## lets create high dimensionality images

```python
img=[]
for i in range(720):
    temp=[]
    for j in range(1280):
        temp.append(0)
    img.append(temp)
img=np.array(img)
plt.imshow(img)
```

## Exporting or Saving the image 

```python
lst=[]
for i in range(720):
    temp=[]
    for j in range(1280):
        temp.append(0)
    lst.append(temp)
lst=np.array(lst)
plt.imshow(lst)

cv.imwrite('img.png',lst)
```

## Creating a RED Image

```python
red=np.array([[[255,0,0],[255,0,0],[255,0,0]],
             [[255,0,0],[255,0,0],[255,0,0]],
             [[255,0,0],[255,0,0],[255,0,0]]])

plt.imshow(red)
```

## RGB Image

```python
rgb=np.array([[[255,0,0],[255,0,0],[255,0,0]],
             [[0,255,0],[0,255,0],[0,255,0]],
             [[0,0,255],[0,0,255],[0,0,255]]])

plt.imshow(rgb)
```

## Black & White Chess Board

## Multicolored Image

```python
qwe=np.array([[[70,0,0],[150,0,0],[255,0,0]],
             [[0,70,0],[0,150,0],[0,255,0]],
             [[0,0,70],[0,0,150],[0,0,255]]])

plt.imshow(qwe)
```

```python
img=[]
for i in range(255):
    t=[]
    for j in range(1920):
        t.append([i,0,0])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([i,0,0])
    img.append(t)
for i in range(255):
    t=[]
    for j in range(1920):
        t.append([0,i,0])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([0,i,0])
    img.append(t)
for i in range(255):
    t=[]
    for j in range(1920):
        t.append([0,0,i])
    img.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([0,0,i])
    img.append(t)
    
    
plt.imshow(img)
```
